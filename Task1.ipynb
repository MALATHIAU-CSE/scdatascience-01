```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create a dummy CSV file for demonstration
dummy_data = {'age': [23, 45, 67, 12, 34, 56, 78, 89, 9, 21, 33, 44, 55, 66, 77, 88, 11, 22, 33, 44, 55, 66, 77, 88, 99],
              'population': [100, 150, 200, 50, 120, 180, 250, 300, 30, 90, 110, 130, 160, 190, 220, 280, 40, 80, 100, 120, 140, 170, 200, 230, 260]}
df_dummy = pd.DataFrame(dummy_data)
dummy_file_path = 'population_data.csv'
df_dummy.to_csv(dummy_file_path, index=False)

# Assuming the file is named 'population_data.csv'
df = pd.read_csv(dummy_file_path)


# Using Matplotlib
plt.hist(df['age'], bins=range(0, 101, 10), edgecolor='black')
plt.xlabel('Age Group')
plt.ylabel('Population Count')
plt.title('Population Distribution by Age (Matplotlib)')
plt.show()

# Using Seaborn (often produces more aesthetically pleasing charts)
sns.histplot(data=df, x='age', bins=range(0, 101, 10))
plt.xlabel('Age Group')
plt.ylabel('Population Count')
plt.title('Population Distribution by Age (Seaborn)')
plt.show()

# Clean up the dummy file
os.remove(dummy_file_path)
```
